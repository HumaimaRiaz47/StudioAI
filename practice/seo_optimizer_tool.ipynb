```python

from typing import List

from pydantic import BaseModel, Field

from langchain_core.messages import (
    SystemMessage,
    HumanMessage,
)

from langchain_ollama import ChatOllama
from schemas import SEOOptimizerOutput
```

```python

llm = ChatOllama(
    model="qwen2.5:7b",
    temperature=0.4,
)
```

```python
# SYSTEM PROMPT

SEO_OPTIMIZER_SYSTEM_PROMPT = """
You are an elite SEO content optimization specialist.

Your job is to optimize existing content for:

- search engine visibility
- readability
- keyword optimization
- engagement
- user search intent

You are NOT writing from scratch.

You are improving already-written content.

===========================================================
YOUR TASKS
===========================================================

1. Improve SEO title
2. Improve heading structure
3. Improve keyword placement
4. Improve readability
5. Add SEO-friendly formatting
6. Add semantic keyword variations
7. Improve introduction hook
8. Improve conclusion
9. Improve search intent alignment
10. Keep content natural and human-readable

===========================================================
IMPORTANT RULES
===========================================================

- Do NOT keyword stuff
- Keep content natural
- Preserve original meaning
- Maintain professional tone
- Keep markdown formatting
- Use clear headings
- Improve flow and structure

===========================================================
SEO BEST PRACTICES
===========================================================

- Use descriptive H1/H2/H3 headings
- Optimize title clarity
- Improve scannability
- Use concise paragraphs
- Add bullet points where useful
- Improve semantic relevance
- Optimize for user engagement

===========================================================
OUTPUT FORMAT
===========================================================

Return ONLY valid JSON.

Example:

{
  "seo_title": "Best AI Marketing Tools in 2026",
  "meta_description": "Discover the top AI marketing tools transforming businesses in 2026.",
  "focus_keywords": [
    "AI marketing tools",
    "marketing automation",
    "AI business tools"
  ],
  "optimized_content": "..."
}

===========================================================
FINAL INSTRUCTION
===========================================================

Act like a professional SEO strategist working for a
high-traffic enterprise content platform.
"""
```

```python
def optimize_seo(
    article: str,
    topic: str,
    target_audience: str = "general audience",
):
    
    structured_llm = llm.with_structured_output(
        SEOOptimizerOutput
    )

    response = structured_llm.invoke(
        [
            SystemMessage(
                content=SEO_OPTIMIZER_SYSTEM_PROMPT
            ),
            HumanMessage(
                content=f"""
TOPIC:
{topic}

TARGET AUDIENCE:
{target_audience}

ARTICLE:
{article}

IMPORTANT:
- Optimize this content for SEO
- Improve readability
- Improve structure
- Improve keyword relevance
- Keep the content natural
"""
            )
        ]
    )

    return response.model_dump()
```
